```python
from PIL import Image
import matplotlib.pyplot as plt
import os
```

```python
pil_im = Image.open('shu.jpg').convert('L')
pil_im2 = Image.open('shu.jpg')
pil_im.save('greenShu.jpg')
```

```python
imgplot = plt.imshow(pil_im)
plt.show()
```

```python
imgplot = plt.imshow(pil_im2)
plt.show()
```

```python
filelist = ['shu.jpg']
for infile in filelist:
    outfile = os.path.splitext(infile)[0] + "2.jpg"
    print(outfile)
    if infile != outfile:
        try:
            Image.open(infile).save(outfile)
            
        except IOError:
            print("...", infile)
```

```python
box = (100,100,400,400)
region = pil_im.crop(box)
region.save('Solem/shuReg.jpg')
imgplot = plt.imshow(region)
plt.show()
```

```python
def paste(region):
    region = region.transpose(Image.ROTATE_180)
    pil_im.paste(region,box)

paste(region)

out = pil_im.rotate(45)

imgplot = plt.imshow(pil_im)
plt.show()
imgplot = plt.imshow(out)
plt.show()
```
